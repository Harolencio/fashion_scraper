```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.get('https://us.fashionnetwork.com/news/s?page=1&categs[0]=60&categs[1]=70&sectors[0]=1&sectors[1]=5&sectors[2]=6&sectors[3]=7&sectors[4]=8&sectors[5]=9&sectors[6]=10&sectors[7]=11&sectors[8]=12&sectors[9]=13&sectors[10]=14&sectors[11]=31')
time.sleep(5)
main = browser.find_element("xpath",'/html/body/div[2]/div[2]/div/div[2]/div/div/div[2]/div[2]/div/ul')
```

```python
articles = main.find_elements(By.CLASS_NAME, 'list-ui__item')
links = []
for article in articles:
    header = article.find_element(By.CLASS_NAME, "list-ui__faux-link")
    link = header.get_attribute('href')
    links.append(link)

```

```python
links
```

```python
main = browser.find_element(By.TAG_NAME, 'article')
fechas = main.find_element(By.CLASS_NAME, 'time-ago')
fecha = fechas.find_element(By.CLASS_NAME, 'time-ago__text')

fecha.text
```

```python
new_items = []
#<span class="time-ago__text">Oct 10, 2022</span>

for link in links:
    browser.get(link)
    record = dict()
    main = browser.find_element(By.TAG_NAME, 'article')
    fechas = main.find_element(By.CLASS_NAME, 'time-ago')
    fecha = fechas.find_element(By.CLASS_NAME, 'time-ago__text')
    
    parrafos = main.find_elements(By.TAG_NAME, 'p')
    #print(parrafo)
    texto = ''
    for i in range(len(parrafos)):
        #texto = ''
        parrafo = parrafos[i].text
        print(parrafo)
        texto += parrafo
    record['Published Date'] = fecha.text        
    record["corpus_text"] = texto
    new_items.append(record)
```

```python
for i in range(len(parrafos)):
    print(i)

#len(parrafos)
```

```python
df = pd.DataFrame(new_items)
df
```
